## Modell 1

## Modell 2

All innstråling vil ikke jorda ta opp. Jorda vil reflektere noe av innstrålingen, målet for refleksjon av en flate kalles albedo.
Vi bruker albedo som en faktor inn i beregningen av temperatur. Vi antar at temperaturen på jorden er konstant. 


vi bruker energiprinsippet og  Stefan-Boltzmanns lov og får at 
siden Energi_inn = energi_ut

Ks⋅(1−albedo)⋅πRE2 = σT4⋅4πRE2

og kan omskrive det som formelen T=⋅((1−albedo)/4σ)**1/4

## Modell 3

Atmosfæren er et veldig komplisert naturfenomen, og vi må forenkle dette for å ikke bruke for mye tid på å lage en modell av dette. 

Vi tar for oss tre antagelser om atmosfæren, 
1. atmosfæren har konstant tempratur, hele atmosfære har den samme temperaturen
2. atmofæren all stråling fra solen treffer jordoverflaten og går "igjennom" atmosfæren
3. atmosfæren tar imot all stråling fra jorden

Her er en forenklet modell som vi kan ta for oss

atmosfæren er tenkt som en rektangel, vist i grå farge. Stefan-Boltzmanns lov gjelder for både jordkloden og atmosfæren. 
Vi kalkulerer energien til atmosfæren ved å bruke stefan-boltzmans lov, og ser hvordan atmosfæren påvirker temperaturen. 

Vi bruker også energiprinsippet, som er en lov om at energien er konstant, altså energi kan ikke plutselig skapes eller forsvinne.

Energi_inn = energi_ut

Vi kan skrive om dette uttrykket til 
Varmestråling * (1-albedo_jorda) = o*T^4 

denne likningen dekker pil 1 og 3, fordi vi tar for oss varmestrålingen som kommer til jorda og hvor mye jorda sender tilbake, 

Vi trenger en til likning for pil 2 siden vi må huske at noe av strålingen vil bli sendt tilbake til jorden. 

Likningen blir da Varmestråling * (1-albedo_jorda)+o*T^4 = o*T^4s

Vi setter likning 1 til 2 og får

Ts = ((varmestråling*(1-albedo_jorda)+varmestråling*(1-albedo_jorda))/o)**1/4



## Modell 4

Vi må endre på modellen vår, fordi vi får for mye innstråling inn på jorda. Vi har antatt at all varmestråling går igjennom atmosfæren, men det er ikke tilfelle. 
Atmosfæren reflekterer også noe av varmestrålingen. Vi finner ut at albedoen til atmosæfren er ca 20%. Fra Modell 3 finner vi ut av at ikke alt av varmestrålingen sola sender ut treffer jorden. så vi regner med vår nye konstant som er ca 340. (som vi kommer fram til i modell 3 drøfting)
dette er fordi sola vil ikke treffe jorda vertikalt rett på jorda, siden jorda ikke er et rekangel og jorda har en vinkel på rundt 21-24 grader. 

Vi setter inn albedoen til atmosfæren til formelen og får denne formelen

Ts = ((varmestråling(1-albedo_jorda-albedo_atmosfæren)+varmestråling(1-albedo_jorda-albedo_atmosfæren))/o)**1/4

## Modell 1

```python
import math as math

#gjennomsnittlig varmestråling som treffer vertikalt på jordklodens atmosfære. 

#Stefan-Boltzmanns naturlov

o = 5.67e-8  #enhet W/(m^2*K^4)
tempratur_sol = 5778 #kelvin
H_sol = o*(tempratur_sol**4) #enhet W/m^2, er kraften fra sola per volum. 

R_sol =  695e6 #enhet m, er radiusen til sola
distansen_sol_jord = 149600000e3 #m, distansen fra sola til jorda. 


Varmestråling = (R_sol**2)/(distansen_sol_jord**2)*H_sol
print(Varmestråling)
```

## Modell 2

```python
albedo_jorda = 0.30

temperatur = ((Varmestråling*(1-albedo_jorda))/(4*o))**(1/4)
temperatur = temperatur - 273 #gjør om temperaturen fra kelvin til celsius. 

print (temperatur)

```

## Modell 3

```python
temperatur =((Varmestråling*(1-albedo_jorda)+Varmestråling*(1-albedo_jorda))/o)**(1/4)
temperatur = temperatur - 273 # gjør om fra kelvin til celsius
print(temperatur) 

#Vi ser at temperaturen er veldig høy, vi finner ny varmestråling konstant, se 4. diskusjon for begrunnelse. 

Varmestråling_ny = Varmestråling/4 #W/M^2
temperatur_ny =((Varmestråling_ny*(1-albedo_jorda)+Varmestråling_ny*(1-albedo_jorda))/o)**(1/4)
temperatur_ny = temperatur_ny - 273 # gjør om fra kelvin til celsius
print(temperatur_ny) 
```
