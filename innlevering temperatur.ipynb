### Temperatur

```python
from pylab import *

data = loadtxt('temperatur.txt', skiprows=1)


tid = data [:,0]
grader = data[:,1] 

plot(tid, grader,)

xlabel('tid (min)')
ylabel('Grader')
show()

```

### Regresjon av temperatur

```python
a, b, c = polyfit(tid, grader, 2)

def modell(x):
    return a*x**2 + b*x + c

x = linspace(0,500,10000)
y = modell(x)
plot(x,y)
plot(tid,grader,linestyle='',marker='.',label='Punkter')
legend()
show()
```
